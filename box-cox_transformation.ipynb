****Box-cox Transformation****

ðŸ‘‰ This technique is particularly useful when you need to stabilize variance and make data more normal distribution like. It can handle both +ve and -ve values by finding an optimal lambda parameter.


**Use case**

ðŸ‘‰ It is used to transform non-normal dependent variables into a normal shape. It can be applied to data that is strictly +ve and helps stabilize variance and make the data more normally distributed.

****Characteristics****

[1]. **Felxible power transformation**:
    It can handle variety of data distribution by finding the optimal lambda parameter.

[2]. **Handles positive values**:
    It only supports/Work with +ve values only.

[3]. **Normalizing effects**:
    It can make data more normally distributed, reducing skewness & stabilizing variance.

[4]. **Optimal lambda**:
    The transformation is parameterized by lambda value, which is estimated to best normalize the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = np.random.exponential(scale=2, size=1000)
boxcox_data = stats.boxcox(data)
```

```python
def visulize(original_data, transformed_data, title='Transfomation'):
    sns.set(style='whitegrid', palette='muted')
    fig, axes = plt.subplots(1,2, figsize=(14,6))
    
    # Normal distribution
    sns.histplot(original_data, bins=30, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title('Data distribution without transformation')
    axes[0].set_xlabel('Data')
    axes[0].set_ylabel('Frequency')
    
    # Transformed distribution
    sns.histplot(transformed_data, bins=30, kde=True, ax=axes[1], color='salmon')
    axes[1].set_title('Data distribution with transformation')
    axes[1].set_xlabel('Data')
    axes[1].set_ylabel('Frequency')
    
    plt.suptitle(title, fontsize=13)
    plt.tight_layout()
    plt.show()
```

```python
visulize(data, boxcox_data[0], 'Box-cox transformation')
```

```python
sample_data_a = np.arange(1,10001,1)
visulize(sample_data_a, stats.boxcox(sample_data_a)[0], 'Box-cox transformation')
```
