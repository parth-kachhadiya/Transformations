**Log transformation**

Log transformation is used when you have data with wide range of values, it helps to stabilize the variance and make
the data more normally distributed, It's preticularly useful for positive skewed data.

Mathematical representation : 

Positively skewed(Long tail on right side) = Right skewed, 
Negatively skewed(Long tail on left side) = Left skewed

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.arange(1, 10001, 1)
log_data = np.log(data)
```

```python
data
```

```python
log_data
```

```python
def visulize(original_data, transformed_data, title='Transfomation'):
    sns.set(style='whitegrid', palette='muted')
    fig, axes = plt.subplots(1,2, figsize=(14,6))
    
    # Normal distribution
    sns.histplot(original_data, bins=30, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title('Data distribution without transformation')
    axes[0].set_xlabel('Data')
    axes[0].set_ylabel('Frequency')
    
    # Transformed distribution
    sns.histplot(transformed_data, bins=30, kde=True, ax=axes[1], color='salmon')
    axes[1].set_title('Data distribution with transformation')
    axes[1].set_xlabel('Data')
    axes[1].set_ylabel('Frequency')
    
    plt.suptitle(title, fontsize=13)
    plt.tight_layout()
    plt.show()
```

```python
visulize(data, log_data, 'Log transformation')
```

```python
sample_data_a = np.arange(1,1001, 10)
visulize(sample_data_a, np.log(sample_data_a), 'Log transformation') 
```

```python
# Normally distributed data

np.random.seed(0)
normal_distribution = np.random.normal(loc=50, scale=10, size=1000)
visulize(normal_distribution, np.log(normal_distribution), 'Log transformation')
```

```python
# Positively skewed data

positive_skewed = np.random.exponential(scale=2, size=1000)
visulize(positive_skewed, np.log(positive_skewed), 'Log transformation')
```

```python
negative_skewed = -np.random.exponential(scale=2, size=1000)
visulize(negative_skewed, np.log(-negative_skewed), 'Log transformation')
```

****Characteristics of Log Transformation****

[1]. **Variance stabilization** :
    Log transformation helps stabilize the variance in the data, especially when the data spans several oders of magnitude.
    This makes it easier to work with the data and can improve the performance of statistical models.

[2]. **Reduction skewness** :
    Log transformation can reduce positive skewness in the data, making the distribution more symmetric or even negatively skewed. 
    This is particualarly useful when dealing with right-skewed data.

[3]. **Normalization** :
    By compressing the range of data values, log transformation can make the data distribution more normal (bell-shaped),
    Which is often an assumption for many statistical models.

[4]. **Handle exponential growth** :
    It is effective in handling data with exponential growth patterns. For ex. financial growth, financial data, population growth,
    and other naturally occuring phenomena that follow exponential trends.

[5]. **Monotonic transformation** :
    It is monotonic, meaning that 'It preserves the order of data points'. If one data point is larger than another, it will remain
    larger after transformation.

[6]. **Data scaling** :
    It helps to scale down the large values, making the data easier to visualize and interpret. It compresses the range of high values
    more than the range of low values.

[7]. **Applicability** :
    It can only be applicable on positive values. If the data contains zeros or negative values, they need to e shifted to positive
    values before transformation.

[8]. **Impact on interpretation** :
    While log transformation makes the data easier to model, it also changes the interpretation. After transformation, the differences are
    on a multiplicative scale rather than an additive scale.

[9]. **Sencitivity to small values** :
    It can be sencitive to small values. It drametically compresses lower values, which might not always be desirable.

[10]. **Non-linearity** :
    The transformation is non-linear, which can help linearize relationships between variables that exhibit non-linear patterns in their raw form.
